```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
req = requests.get('https://www.passportindex.org/byRank.php')
```

```python
soup = BeautifulSoup(req.content)
```

```python
print(soup)
```

```python
country_name_html = [i for i in soup.find_all(class_='name_country')]
country_name_html
```

```python
country_name = [country.text for country in country_name_html]
country_name
```

```python
VFS_html = [i for i in soup.find_all(class_='name_rank')]
VFS_html
```

```python
VFS = [vfs.text for vfs in VFS_html]
VFS
```

```python
VF_html = [i for i in soup.find_all(class_='rank vf')]
#print(VF_html)
VF = [vf.text for vf in VF_html]
VF
```

```python
VOA_html = [i for i in soup.find_all(class_='rank voa')]
#print(VF_html)
VOA = [VOA.text for VOA in VOA_html]
VOA
```

```python
VR_html = [i for i in soup.find_all(class_='rank vr')]
#print(VF_html)
VR = [VR.text for VR in VR_html]
VR
```

```python
dict1 = {'Country':country_name, 'VFS':VFS, 'VF':VF, 'VOA':VOA, 'VR':VR}
dict1
```

```python
df = pd.DataFrame(dict1)
df.head()
```
